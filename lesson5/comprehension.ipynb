### list comprehension

```python
# example 1
numbers = [2, 3, 4, 5, 6]
a_squares=[]
for number in numbers :
    a_squares.append(number**2)
print(a_squares) 
```

```python
# example 1.2
numbers = [2,3,4,5,6]
a_square=[number**2 for number in numbers]
print(a_square)
```

```python
# example 2
numbers = [2,3,4,5,6]    
a_even_square = []
for number in numbers: 
    if number % 2 == 0:
        a_even_square.append(number**2)
print(a_even_square)
```

```python
# example2.2
numbers =[2,3,4,5,6]
a_even_square=[ number**2  for number in numbers if number %2 == 0]
print(a_even_square)
```

```python
# example 2.3
numbers =[2,3,4,5,6]
a_even_square=[(number, number**2 ) for number in numbers if number %2 == 0]
print(a_even_square)
```

```python
# example 3
numbers =[2,3,4,5,6]
a_even_square=[print(number)  for number in numbers if number %2 == 0]
print(a_even_square)
```

```python
# example 6
matrix= [[1,2,3],[4,5,6], [7,8,9]]
flattened= []
for row in matrix:
    for val in row:
        flattened.append(val)
print(flattened)
```

```python
 # example 6.2
matrix= [[1,2,3],[4,5,6], [7,8,9]]
flattened= [val for row in matrix for val in row]
print(flattened)
```

## dictionary comprehension

```python
keys = ['a', 'b', 'c', 'd', 'e']
values = [1, 2, 3, 4, 5]
dict = {k: v for k, v in zip(keys, values)}
print(dict)
```
