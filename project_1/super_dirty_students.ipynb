```python
import pandas as pd
import re

def clean_age(value):

    if pd.isna(value):
        return pd.NA
    
    value = str(value).lower().strip()
    
 
    if value.isdigit():
        return int(value)

    numbers = re.findall(r"\d+", value)
    if numbers:
        return int(numbers[0])
    

    word_to_num = {
        "eighteen": 18,
        "nineteen": 19,
        "twenty": 20,
        "twenty one": 21,
        "twenty two": 22
    }
    
    if value in word_to_num:
        return word_to_num[value]
    
    return pd.NA


df["age"] = df["age"].apply(clean_age)
df["age"] = df["age"].astype("Int64") 
```

```python
import pandas as pd
import numpy as np

def parse_mixed_datetime(x):
    if pd.isna(x):
        return pd.NaT
    
    x = str(x).strip()
    
   
    if x.isdigit() and len(x) == 10:
        return pd.to_datetime(int(x), unit='s', errors='coerce')
    
   
    if "-" in x or "T" in x or "." in x:  
        try:
            return pd.to_datetime(x, errors='coerce', dayfirst=False)
        except:
            return pd.NaT
    
   
    if "/" in x:
        try:
            return pd.to_datetime(x, errors='coerce', dayfirst=True)
        except:
            return pd.NaT
  
    try:
        return pd.to_datetime(x, errors='coerce')
    except:
        return pd.NaT


date_cols = ['date_of_join', 'event_time']
for col in date_cols:
    df[col] = df[col].apply(parse_mixed_datetime)
```

```python
def format_uz_phone(number):
    if str(number).startswith('998'):
        return '+'+ number
    elif str(number).startswith(('9', '7', '5')):
        return '+998' + str(number) 
    else:
        return None 

df['phone'] = df['phone'].apply(format_uz_phone)     
```

```python
import pandas as pd
import json
import ast
from pandas import json_normalize

# 1️ Duplicate columnlarni olib tashlash
df = df.loc[:, ~df.columns.duplicated()]

# 2️ Safe parse: JSON yoki Python dict stringni dict ga aylantirish
def smart_parse(x):
    if isinstance(x, dict):
        return x
    if pd.isnull(x):
        return {}
    if isinstance(x, str):
        try:
            return json.loads(x)
        except:
            try:
                return ast.literal_eval(x)
            except:
                return {}
    return {}

df['profile_json'] = df['profile_json'].apply(smart_parse)

# 3️ Avtomatik flatten qilish (first level)
json_df = json_normalize(df['profile_json'])
df = pd.concat([df.drop(columns=['profile_json']), json_df], axis=1)

# 4️ Hobbies va devices listlarini stringga aylantirish
list_columns = ['hobbies']

for col in list_columns:
    if col in df.columns:
        df[col] = df[col].apply(
            lambda x: ', '.join(x) if isinstance(x, list) and all(isinstance(i, str) for i in x) else x
        )

# 5️ Skills nested dict ichidagi tech ni alohida column qilish
if 'skills' in df.columns:
    df['python_skill'] = df['skills'].apply(lambda x: x['tech']['python'] if isinstance(x, dict) and 'tech' in x else None)
    df['excel_skill']  = df['skills'].apply(lambda x: x['tech']['excel']  if isinstance(x, dict) and 'tech' in x else None)
    df['sql_skill']    = df['skills'].apply(lambda x: x['tech']['sql']    if isinstance(x, dict) and 'tech' in x else None)
    df = df.drop(columns=['skills'])  # eski nested dict ustunni o'chiramiz

# 6️ Family nested object flatten
if 'family' in df.columns:
    df['siblings'] = df['family'].apply(lambda x: x.get('siblings') if isinstance(x, dict) else None)
    df['married']  = df['family'].apply(lambda x: x.get('married') if isinstance(x, dict) else None)
    df = df.drop(columns=['family'])

 #Devices list of dict flatten (misol: birinchi element)
if 'devices' in df.columns:
    df['device_type']  = df['devices'].apply(lambda x: x[0]['type'] if isinstance(x, list) and len(x) > 0 else None)
    df['device_brand'] = df['devices'].apply(lambda x: x[0]['brand'] if isinstance(x, list) and len(x) > 0 else None)
    df['device_year']  = df['devices'].apply(lambda x: x[0]['year'] if isinstance(x, list) and len(x) > 0 else None)
    df = df.drop(columns=['devices'])

# 8️ Natijani tekshirish
print(df.head())
```

```python
if 'address_row' in df.columns:
    
    df['addr_city'] = df['address_row'].apply(
        lambda x: x.split(',')[0].strip() if isinstance(x, str) else None
    )
    
    df['addr_district'] = df['address_row'].apply(
        lambda x: x.split(',')[1].strip() if isinstance(x, str) and len(x.split(',')) > 1 else None
    )
    
    df['addr_postal'] = df['address_row'].apply(
        lambda x: x.split(',')[2].strip() if isinstance(x, str) and len(x.split(',')) > 2 else None
    )
```

```python
import pandas as pd




hashable_cols = [col for col in df.columns 
                 if df[col].apply(lambda x: isinstance(x, (int, float, str, bool, type(None)))).all()]

print("Hashable ustunlar tekshirilmoqda:", hashable_cols)

duplicate_count = df.duplicated(subset=hashable_cols).sum()
print("Duplicate qatorlar soni:", duplicate_count)

df = df.drop_duplicates(subset=hashable_cols)


missing_info = df.isnull().sum()
print("\nMissing values har bir ustunda:")
print(missing_info)


important_cols = ['Age', 'GPA', 'score'] 

for col in important_cols:
    if col in df.columns:
       
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col] = df[col].fillna(df[col].mean())
       
        elif pd.api.types.is_string_dtype(df[col]):
            df[col] = df[col].fillna('Unknown')


df = df.dropna()  


print("\nDuplicate va missing values tozalangan DataFrame:")
print(df.info())
```
