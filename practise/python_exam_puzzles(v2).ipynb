```python
def has_duplicates(nums):
    return len(nums)!=len(set(nums))

print(has_duplicates([1,2,3,1]))
print(has_duplicates([1,2,3,4]))
```

```python
recipe = pd.DataFrame({
    "recipe_name": ["hotdog","hotdog","pilov","pilov","pilov","pizza","pizza","pizza"],
    "ingredient_id": [1,2,3,4,5,6,7,8],
    "ingredient_name": ["sausage","bread","rice","carrot","meat","flour","tomato","cheese"]
})

fridge = pd.DataFrame({
    "ingredient_id": [1,2,3,4,6],
    "ingredient_name": ["sausage","bread","rice","carrot","flour"]
})

merged=recipe.merge(fridge, on="ingredient_id", how="left", indicator=True)
available_recipes=merged.groupby("recipe_name")["_merge"].apply(lambda x: all(x=="both"))
result=available_recipes[available_recipes].index.tolist()

print(result)
```

```python
def reverse_letters_digits(s):
    chars=[c for c in s if c.isalnum()]
    chars.reverse()
    result=[]
    for c in s:
        if c.isalnum():
            result.append(chars.pop(0))
        else:
            result.append(c)
    return"".join(result)
        
print(reverse_letters_digits("a!1b@2c"))
```

```python
def digits_sum(s):
    return sum(int(c)for c in s if c.isdigit())
```

```python
def is_prime(n):
    if n<=1:
        return False
    if n==2:
        return True
    for i in range(3, int(n**0.5)+1, 2):
        if n % i == 0:
            return False
    return True
```

```python
import sqlite3
conn=sqlite3.connect("example.db")
cur=conn.cursor()
cur.execute("""
               create table roster2(
               name text,
               species text,
               age integer
               )
               """)
cur.executemany("insert into roster2 values(?,?,?)", [
    ("Charlie","dog",3),
    ("Mittens","cat",2),
    ("Buddy","dog",5),
    ("Whiskers","cat",4)])

conn.commit()
for row in cur.execute("select name, age, species from roster2 where species='cat'"):
    print(row)
```
