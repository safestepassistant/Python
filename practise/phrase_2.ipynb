```python
def max_num(a,b,c):
    if a>=b and a>=c:
        return a
    elif b>=a and b>=c:
        return b
    else:
        return c
    
print (max_num(9,1,3))
print (max_num(7,8,5))
```

```python
def score_calculate (score):
    if score>100 or score<0:
        return("sorry,there is no such score. please enter your actual score")
    elif  90<=score<=100:
        return("Super, your level is 'A'")
    elif  80<=score<=89:
        return("your level is 'B'")
    elif  70<=score<=79:
        return("your level is 'C'")
    elif 60<=score<=69:
        return("your level is 'D'")
    else :
        return("your score is too low and you failed") 
print (score_calculate (101))
print (score_calculate (-1))
print (score_calculate (75))
print (score_calculate (55))
print( score_calculate (93))
```

```python
def sum_num(a):
    sumofall=0
    for i in range(0,a+1):
        sumofall+=i
    return sumofall
print(sum_num(5))
print(sum_num(9))
```

```python
def password_checker(password):
    if len(password)>=8 and any(p.isdigit() for p in password) and any(p.isalpha()for p in password):
        return "strong"
    else:
        return "weak"
print( password_checker('12345678'))
print( password_checker('onetwothree'))
print( password_checker('onetwothree1'))
```

```python
def calculator(a,b,operator):

    if operator =='+':
         return(a+b)
    elif operator=='-':
        return(a-b)
    elif operator=='*':
        return(a*b)
    elif operator=='/':
        if b!=0:
            return(a/b)
        else:
            return("division by zero is not allowed")
    else:
        return("invalid operator")
    
print (calculator(5,6,'+'))
print (calculator(5,8,'*'))
print (calculator(6,2,'/'))
```
