```python
def digit_sum(k):
    return sum(int(digit) for digit in str(k))
print(digit_sum(1234567890))
```

```python
def is_prime(n):
    if n<=1:
        return False
    if n==2:
        return True
    for i in range (3,int(n**0.5)+1,2):
        if n%i==0:
            return False
        return True
    
print(is_prime(29))
print(is_prime(21))
```

```python
import sqlite3

connection=sqlite3.connect("example.db")
cur=connection.cursor()
cur.execute("""
    create table if not exists roster(
    name text,
    species text,
    age integer)
            """)

cur.executemany("insert into roster values (?,?,?)",[
    ("Fido","dog",4),
    ("Whiskers","cat",7),
    ("Rover","dog",2)
])
connection.commit()
for row in cur.execute("select name ,age from roster where species='rover'"):
    print(row)
```

```python
import pandas as pd
df=pd.read_csv("customer_orders.csv")

order_counts=df.groupby("customerid").size()
customers_20plus=order_counts[order_counts>=20].index

avg_price=df.groupby("costumerid")['price'].mean()
high_spenders=avg_price[avg_price>120].index

print("Customers with 20 or more orders:", customers_20plus.tolist())
print("Customers with high average order price above $120:", high_spenders.tolist())
```
