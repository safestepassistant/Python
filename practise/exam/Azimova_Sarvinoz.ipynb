```python
#problem_2
# Print all prime numbers within a range (25, 50)
# Use a function to complete the task.
def is_prime(num):
    if num < 2:
        return False
    if num % 2 == 0:
        return num == 2
    for i in range(3, int(num**0.5) + 1, 2):
        if num % i == 0:
            return False
    return True

for n in range(25, 51):
    if is_prime(n):
        print(n)
```

```python
# Using employee_data.csv file.
# task 1. Create 4 separate csv files for 4 different countries
# task 2. Find the Most Experienced Employee in Each Country
# task 3. Find Employees Earning Above Their Country’s Average Salary
import csv
from collections import defaultdict

employees_by_country = defaultdict(list)
with open("employee_data.csv", newline="", encoding="utf-8") as csv_file:
    reader = csv.DictReader(csv_file)
    for row in reader:
        country = row.get("Country") or row.get("country") or "Unknown"
        employees_by_country[country].append(row)

# Task 1: Create up to 4 separate CSV files (one per country)
for i, (country, rows) in enumerate(employees_by_country.items(), start=1):
    if i > 4:
        break
    out_name = f"employee_data_{country}.csv"
    with open(out_name, "w", newline="", encoding="utf-8") as out:
        writer = csv.DictWriter(out, fieldnames=reader.fieldnames)
        writer.writeheader()
        writer.writerows(rows)

# Task 2: Most experienced employee in each country
for country, rows in employees_by_country.items():
    most_experienced = max(rows, key=lambda r: float(r.get("YearsExperience") or 0))
    print(f"{country}: {most_experienced.get('Name')} ({most_experienced.get('YearsExperience')} years)")

# Task 3: Employees earning above their country’s average salary
for country, rows in employees_by_country.items():
    salaries = [float(r.get("Salary") or 0) for r in rows]
    if not salaries:
        continue
    avg = sum(salaries) / len(salaries)
    above_avg = [r for r in rows if float(r.get("Salary") or 0) > avg]
    print(f"\n{country} - avg salary: {avg:.2f}")
    for r in above_avg:
        print(f"  {r.get('Name')} ({r.get('Salary')})")
```
