```python
#Using student.json file,
#create CSV file called student.csv with following columns like ID,Name,Age
import json
import csv
with open("student.json")as json_file:
    data =json.load(json_file)
with open("student.csv","w",newline="")as csv_file:
    writer=csv.writer(csv_file)
    writer.writerow(["ID","Name","Age"])
    for student in data["students"]:
        writer.writerow([student["id"],student["name"],student["age"]])

        
```
