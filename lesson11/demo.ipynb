```python
import sqlite3
```

### 1 method

```python
connection=sqlite3.connect("test.db")
query="select datetime('now', 'localtime')"
cursor=connection.cursor()
time = cursor.execute(query)

time = time.fetchone()[0]
print(time)
connection.close()
```

```python
with sqlite3.connect("test.db") as connection:
    cursor=connection.cursor()
    query="select datetime('now', 'localtime')"
    data=cursor.execute(query)
data.fetchone()
data.fetchmany(1)
data.fetchall()
connection.close()                  
```

### 4 method

```python
sql = """ 
    drop table if exists students3;

    create table students3(fname text, lname text, age int);

    insert into students3 values
    ('ali','vali', 21),
    ('ali','vali', 21),
    ('ali','vali', 21);
"""
```

```python
with sqlite3.connect("students3.db")as connection:
    cursor=connection.cursor()
    cursor.executescript(sql)
```

### select

```python
with sqlite3.connect("students3.db") as connection:
    cursor = connection.cursor()
    query= "select * from students3"
    data = cursor.execute(query)
data.fetchall()
```

```python
with sqlite3.connect("students3.db") as connection:
    cursor = connection.cursor()
    query= "select * from students3 order by age "
    data = cursor.execute(query)
data.fetchmany(2)
```

### 5 method

```python
student_values= (
    ('ali', 'vali', 21),
    ('ali', 'vali', 21),
    ('ali', 'vali', 21)
        )
```

```python
with sqlite3.connect("students3.db") as connection:
    cursor=connection.cursor()
    data= cursor.executemany("insert into students3 values(?,?,?)", student_values)
```

```python
with sqlite3.connect("students3.db")as connection:
    cursor= connection.cursor()
    placeholders=", ".join(["?"]*3)
    data=cursor.executemany(f"insert into students3 values ( {placeholders} )", student_values)
```
